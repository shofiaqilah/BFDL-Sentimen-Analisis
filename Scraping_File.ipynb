```python
import pandas as pd
from google_play_scraper import app, reviews, Sort

# ===== Scraping Data Review Aplikasi BCA Mobile =====
sr = reviews(
    'com.bca',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=50000
)

df = pd.DataFrame(sr[0])
df.to_csv('bcamobile_reviews.csv', index=False)

print(f'Jumlah data diambil : {len(df)}')
print(f'Data berhasil disimpan ke bcamobile_reviews.csv')
```
